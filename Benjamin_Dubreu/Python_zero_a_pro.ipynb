### exercice 1: utilisez le modulo pour savoir si un nombre est pair

```python
import random

nombre_random = random.randint(1, 1000)
# nombre_random % 2 = 0 si le nombre est pair
print(f"le nombre est {nombre_random}, son modulo est {nombre_random % 2}, il est {'pair' if nombre_random %2 == 0 else 'impair'}")
```

### exercice 2 : utilisez 2 **_  jusqu’à trouver celui qui donne 256

```python
x = 2
while x < 256:
    x = 2**x
    print(x)
```

```python
for i in range (1, 10):
    if (2 ** i) == 256:
        print(i)
        break
```

```python
j = 2
k= 0
while k < 256:
    print(j)
    k = 2 ** j
    j += 1
    
# version plus optimisée

i = 0

while 2 ** i != 256:
    i += 1

print(f"2 ** {i} = 256")
```

### exercice 3: transformez 62.137273664980675 miles en kilomètres (1 mile = 1.60934 km)

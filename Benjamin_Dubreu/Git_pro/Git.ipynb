### Une fois notre nouvelle branche créée en local, on la push en ligne. Pour cela il faut faire :

#lie notre branche locale à la branche distante correspondante

git push --set-upstream origin nom_de_la_nouvelle_branche

#explication: origin = raccourci vers l’adresse du dépôt en ligne (GitHub, GitLab, etc.). Quand je clone un dépôt Git crée automatiquement un remote appelé origin qui pointe vers cette URL.
Après avoir fait git push -u origin ma_branche Git mémorise que ma branche locale suit origin/ma_branche . Donc ensuite, un simple git push fonctionne sans préciser origin.
Est-ce que origin peut s’appeler autrement ? oui ! C’est juste un nom. Tu pourrais avoir : upstream, production, monserveur

Par exemple, dans un projet open-source :
origin → ton fork
upstream → le dépôt officiel
origin = le nom du dépôt distant vers lequel tu envoies ton code.

### Mécanisme de merge request

On écrit notre code sur une branche qui n'est pas la main, on le push sur github, on va dans pull request, on va dans "files changed", on check, puis on va dans l'onget discussion, et on merge la pull request

Pour travailler avec git en entreprise:

-> se mettre en local sur la Main ou sur la Develop

-> récupérer la dernière version du code (git pull)

-> créer une branche à partir de là

-> sur cette branche, faire vos modifications de code

-> ajouter ces modifications à l'index

-> faire un commit

-> push votre branche locale sur github (1)

-> aller sur github et créer une pull request de votre branche vers la main (ou la develop s'il y en a une)

-> envoyer votre PR à un collègue pour qu'il vérifie le code

-> si votre collègue accepte les modifications, merger la PR

-> enfin, supprimer notre banche localement avec git branch -d nouvelle_feature

```python
#ordre des commandes

git status
git pull
git checkout -b "update_pipeline_script"
git branch #pour vérifier qu'on est bien sur la bonne branche
git add nom_de_mon_fichier
git commit -m 'description de mes changements'
git push --set-upstream origin nom_de_la_nouvelle_branche
```

### Note

Si vous en avez marre que git vous dise:

"fatal: The current branch master has no upstream branch.",

vous pouvez soit utiliser:

"git push -u origin head"

(mais il faudra le faire à chaque fois)

soit, avec les dernières version de git, modifier une seule fois votre config :

"git config --global push.autoSetupRemote true"

et ensuite vos "git push" fonctionneront tout le temps, même quand il n'y a pas encore d'upstream branch

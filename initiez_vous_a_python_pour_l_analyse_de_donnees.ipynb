# Programmez en orienté objet

La classe peut être vue comme le schéma de construction d'un objet

Une méthode de classe est une fonction disponible uniquement pour les différentes instances de cette classe.

```python
# Quelques méthodes de chaines de caractères
texte = 'je suis un texte'

texte.upper() # retourne l'ensemble du texte en majuscules
texte.capitalize() # retourne le texte en minuscule avec la première lettre en majuscule
texte.replace('j', 'i') # remplace les premiers caractères par les seconds
texte.find('s')

```

Une classe est un plan de construction d'un objet.  
Une variable est une instance de classe, ou un objet.   
Un objet est défini par des attributs.   
Toutes les instances d'une classe ont accès aux mêmes méthodes via la notation .
Une méthode, tout comme une fonction, ne modifie généralement pas l'objet initial.

# Les booléens

```python
# Opérateurs logiques booléens

2 == 2 # -> True
2 == 3 # -> False
4 != 4 # -> False
4!= 5 # -> True
1 < 2 # -> True
1 < 1 # -> False
1 <= 1 # -> True
3 > 4 # -> False
5 > 4 # -> True
5 >= 4 # -> True
```

Ce qui suit if doit aboutir à un booléen. Cela peut-être :

- une valeur  True  ou  False. Par exemple,  if True :  ;

- une variable de type booléen. Par exemple,  if maVariable:  où  maVariable  est un booléen ;

- une expression qui aboutit à une valeur booléenne, comme dans l’exemple vu ci-dessus.

### Opérateurs logiques

opérateur AND
Le résultat final est vrai seulement lorsque toutes les expressions/valeurs sont vraies.

opérateur OR
Le résultat final est vrai lorsqu’au moins une des expressions/valeur est vraie.

opérateur NOT
Cela inverse simplement le résultat de l’expression donnée. Par exemple, le résultat de  not(expression)  est vrai lorsque  expression  est faux.

```python
True and True # True
True and False # False
False and False # False
True or False # True
True or True # True
False or False # False
not(True) # False
not(False) # True
```

```python
# De manière plus complexe 

True and True and True # True
True and True and False # False
True or False or False # True
False or False or False # False
```

De manière plus tricky, comme dans le calcul, certaines opérations sont réalisées en premier.

```python
False or True and True # True
not(False) and True or False # True
```

```python
# Si on utilise des parenthèses on peut changer l'ordre

(True and False) or True # True
not(True and False or not(True)) # True
```

#### L'opérateur "in"

L'opérateur in renvoir True lorsqu'une valeur est trouvée dans une séquence (string ou liste), False sinon

```python
maListe = [4, 2, 3, 2, 10]
maListeDeString = ["a", "b", "c", "d"]
monString = "La météo est vraiment bien aujourd'hui !"

4 in maListe # True
0 in maListe # False
0 in maListeDeString # False
"c" in maListeDeString # True
"e" in maListeDeString # False
"météo" in monString # True
"vraiment" in monString # True
"pluie ?" in monString # False
```

### Boucle for via une valeur itérative entière

range(début, fin, pas), qui va générer une liste de nombres selon trois paramètres :

début  : le premier nombre de la  séquence;

fin  : correspond au dernier nombre de la séquence non inclus. La fonction va générer des nombres de  début  à  fin- 1;

pas  : le pas entre chaque nombre généré.

```python
for i in range(0, 5, 1):
    print(i) # -> affiche de 0 à 4 par pas de 1 (fin - 1)
    
for i in range(0, 5):
    print(i) # -> affiche de 0 à 4 également (le pas par défaut est 1)
    
for i in range(5):
    print(i) # -> affiche de 0 à 4 également (le début par défaut est 0)

for i in range(0, 5, 2):
    print(i) # -> affiche 0, 2 puis 4
```

# Boucle While

## Passer certaines instructions au sein de la boucle. "continue" et "break"

```python
# On peut passer une itération avec 'continue'

for i in range(6):
    # instructions réalisées à chaque itération
    print(i)
    if (i == 2) or (i == 5):
        print("Cas spécial")
        continue 
    # instructions non exécutées si i == 2 or 5 
    #print("i != 2 & i != 5")
    print('lala')

# En fait, continue passe à la prochaine itération de la boucle 
# en ignorant les instructions suivantes de la boucle
```

```python
# On peut aussi décider d'interrompre la boucle, par exemple 
# lorsque l'on cherche un élément particulier dans une liste

panier = ["pomme", "orange", "banane"]

for fruit in panier:
    if fruit == "orange":
        print("J'ai une", fruit, "!")
        break
```

# Manipulez des nombres aléatoires avec le module random

```python
import random
```

```python
for i in range(3):
    print(random.random())
```

```python
#uniform(a, b)  : va générer un float aléatoire compris entre  a  et  b

print('exemple de la fonction uniform : ')
for i in range(4):
    print(random.uniform(5,10))

#randint(a, b)  : va générer un int aléatoire compris entre  a  et  b
print('\nexemple de la fonction randint : ')
for i in range(4):
    print(random.randint(5,10))
```

## Générer une distribution aléatoire selon une distribution donnée

```python
# ex: Gaussienne
# gauss(moyenne, écart-type)

for i in range(10):
    print(random.gauss(0, 1))
```

## Choisir aléatoirement dans une liste

```python
liste  = ['Eva', 'Loris', 'Antoine', 'Jérémy']
for i in range(3):
    print(random.choice(liste))
```

```python
# choisir un échantillon de la liste initale, avec remise

random.choices(liste, k=3)
```

```python
# Pour un échantillon sans remise. On parle de sous-échantillonage

random.sample(liste, 3)
```

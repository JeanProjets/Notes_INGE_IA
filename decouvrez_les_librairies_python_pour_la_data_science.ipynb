# Manipulez les tableaux avec Numpy

```python
!pip install numpy
```

```python
import numpy as np

revenus = [1800, 1500, 2200, 3000, 2172]
revenus_array = np.array(revenus)
revenus_array
```

Si dans la liste de départ il y a des données de types différents, NumPy essaiera de tous les convertir au type le plus général. Par exemple, si un tableau contient des entiers (  int  ) et des nombres décimaux (float ), tous ses éléments seront convertis en nombres décimaux (float ).

```python
n, i, j, p = 3, 3, 3, 3

np.zeros(n) # permet de créer un array rempli de 0, de n éléments ;

np.ones(n) # similaire en tous points à la fonction présentée ci-dessus, à la différence que l’array sera rempli cette fois-ci de 1 ;

np.arange(i, j, p) # permet de créer un array rempli avec une séquence linéaire, qui ira de  i  à  j, par pas de  p  ;

np.linspace(i, j, n) # permet de créer un array de  n  valeurs espacées uniformément entre  i  et  j.
```

```python
# array([1800, 1500, 2200, 3000, 2172])
revenus_array.dtype
```

Les types en NumPy seront toujours présentés via deux informations :

le type (ici  int  ) ;

la précision (ici  32  ).

Cette dernière correspond au nombre de bits sur lesquels sont codés les nombres, définissant ainsi les différentes valeurs possibles. Par exemple, un int8  (entier codé sur 8 bits) pourra prendre des valeurs de -128 à 127 !

#### On accède aux arrays comme on accède aux listes


```python
# pour accéder au 5ème élement
revenus_array[4]

# On peut aussi modifier les valeurs
revenus_array[1] = 1900
```

```python
# Séléctionne l'élément 0, 1, 2
revenus_array[0:3]
```

```python
# Les 3 premiers éléments
print(revenus_array[:3])
```

Nous pouvons accéder à un ensemble d'éléments contigus en combinant []  et  : . La syntaxe suit une règle simple : nom_array[i:j:p] , avec :

 i  : l’indice du début ;

 j  : l’indice de fin ;

p : le pas.

```python
# Un élément sur deux
print(revenus_array[::2])
```

```python
# Si le pas est négatif, le début et la fin du slice sont inversés. On peut utiliser cette propriété pour inverser un tableau.
revenus_array[::-1]
```

# Accédez à plusieurs éléments selon une condition


nom_array[condition de sélection]

```python
revenus_array[revenus_array > 2000]
```

```python
revenus_array[(revenus_array > 500) & (revenus_array < 2800)]
```

#### Cela permet également de modifier des éléments au sein d'un array

```python
revenus_array[revenus_array > 2000] = 0
revenus_array
```

# Méthodes sur les arrays

```python
revenus_array.shape
# (5,) on a array contenant 5 éléments
```

```python
# calculer la moyenne
revenus_array.mean()
```

```python
# calculer la somme
revenus_array.sum()
```

```python
# calculer le maximum (ou le minimum) :
revenus_array.max()
revenus_array.min()
```

```python
# accéder à l’indice de l’élement minimum (ou maximum) :
revenus_array.argmin()
revenus_array.argmax()
```

```python
#ordonner les arrays
revenus_array.sort()
revenus_array
```

### En résumé pour les arrays

On peut sélectionner au sein d’un array :

un élément via son indice avec l’écriture : 

        nom_array[indice]  ;

plusieurs éléments contigus via la syntaxe : 

         nom_array[début:fin:pas]  ;

certains éléments spécifiques via une condition : 

        nom_array[condition]
